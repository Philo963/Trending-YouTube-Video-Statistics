# **Machine learning'**


```python
import pandas as pd

# Load your dataset
# df = pd.read_csv('your_dataset.csv')  # Uncomment and provide the path to your dataset

# Display the first few rows of the DataFrame
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values as needed
df.fillna(0, inplace=True)  # Example: fill missing values with 0
```

```python
# Convert categorical features to numerical
df = pd.get_dummies(df, columns=['category_id'], drop_first=True)

# Select features and target variable
X = df[['views', 'dislikes', 'comments_disabled']]  # Example features
y = df['likes']  # Target variable
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions using the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r_squared:.2f}')
```

```python
import matplotlib.pyplot as plt

# Plot the actual vs. predicted likes
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, label='Predicted Likes', color='red', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)  # Diagonal line
plt.title('Actual vs. Predicted Likes')
plt.xlabel('Actual Likes')
plt.ylabel('Predicted Likes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create and train the random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
rf_y_pred = rf_model.predict(X_test)

# Evaluate the model
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r_squared = r2_score(y_test, rf_y_pred)

print(f'Random Forest Mean Squared Error: {rf_mse:.2f}')
print(f'Random Forest R-squared: {rf_r_squared:.2f}')
```

```python
#Prediction: Linear Regression Model to Predict Likes Based on Views
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Prepare data for prediction
X = df[['views']]
y = df['likes']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Plot the prediction
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, label='Actual Likes', color='blue')
plt.scatter(X_test, y_pred, label='Predicted Likes', color='red', marker='x')
plt.title('Actual vs Predicted Likes based on Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.legend()
plt.show()

# Display model's performance (R-squared)
r_squared = model.score(X_test, y_test)
print(f'R-squared: {r_squared}')
```
